## If happy make the above code into a function

```python
# Take a path and filename, format to desired and return a DataFrame
def process_state(path, filename, empty_before_header=0):
    path_filename = path + filename
    # read csv into df
    state_df = pd.read_csv(path_filename)
    state_df.columns=["DATE", "homeownership_rate"]
    # split filename to get the name of the state and add to the column state
    state_name = filename.split('_')[0]
    state_df ['state'] = state_name
    print(path_filename)
        # renaming the column name to unemployment_rate
        #unemp_df .rename(index = {1: 'unemployment_rate'})
    print(state_df.head(5))
        # formatting the year column to only display year
    state_df ['year']= pd.to_datetime(state_df['DATE'], errors='coerce').dt.year
    state_df .drop(columns = ['DATE'], inplace = True)

    return state_df 
```

## Iterate through all files in a directory and concatenate

```python
import os
import sys

homeownership_df = process_state("","National_Unemployment_rate_1976_2020.csv")

path = "Homeownership/"
empty_rows_above_header = 1

for filename in os.listdir(path):
    if filename[0] != ".":
        
        state_df = process_state(path, filename)
        homeownership_df = pd.concat([homeownership_df, state_df], axis = 0, ignore_index = True)
    
    
homeownership_df.head(60)
```
